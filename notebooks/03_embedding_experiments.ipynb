```python
import sys
sys.path.insert(0, "../src")

from educateai.core.models import Chunk
from educateai.ingestion.embedder import embed_chunks
from educateai.ingestion.store import save_chunks, search

chunks = [
    Chunk(text="Photosynthesis converts sunlight into glucose",
          topic="Biology", source_file="test.pdf", chunk_index=0, char_count=45),
    Chunk(text="The mitochondria produces ATP through respiration",
          topic="Biology", source_file="test.pdf", chunk_index=1, char_count=49),
]

chunks = embed_chunks(chunks)
print(f"Embedded {len(chunks)} chunks")

save_chunks(chunks)
print("Saved to ChromaDB")


# embed a query question and search
from educateai.ingestion.embedder import _model

query = "how is energy released in cells?"
query_vector = _model.encode(query).tolist()

results = search(query_vector, top_k=2)
print(results["documents"])     # text of matching chunks
print(results["distances"])     # how close each match is
print(results["metadatas"])     # topic, source_file etc
```
