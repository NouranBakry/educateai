```python
import sys
sys.path.insert(0, "../src")  # so Python can find educateai package

from educateai.core.models import Chunk
from educateai.ingestion.embedder import embed_chunks

chunk = Chunk(
    text="Photosynthesis converts sunlight into glucose",
    topic="Biology",
    source_file="test.pdf",
    chunk_index=0,
    char_count=45
)

result = embed_chunks([chunk])
print(f"Embedding length: {len(result[0].embedding)}")   # should be 384
print(f"First 5 values: {result[0].embedding[:5]}")       # sample of the vector
print(f"Type: {type(result[0].embedding[0])}")            # should be float
```
